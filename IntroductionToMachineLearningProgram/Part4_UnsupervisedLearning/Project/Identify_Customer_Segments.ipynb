The fourth column of the feature attributes summary (loaded in above as `feat_info`) documents the codes from the data dictionary that indicate missing or unknown data.
```python
# Identify missing or unknown data values and convert them to NaNs
azdias.isna().sum()
```

```python
# Functional approach
def missing_dict_fun(df=feat_info, key_column='attribute', value_column='missing_or_unknown'):
    """
    Function that takes in a dataframe and returns a dictionary of the missing values indicated for an attribute
    
    : param df: Pandas dataframe
    : param key_column: name of column whose values will associate with the missing value codes
    : param value_column: name of column with the missing value codes
    
    : return missing_dict: dictionary of column names as keys and missing value codes as values
    """
    count = 0
    mt_dict = {}
    
    for attribute in df.loc[:, key_column]:
        mt_dict[attribute] = df[df[key_column] == attribute][value_column][count]
        if len(mt_dict[attribute]) > 2:
            mt_dict[attribute] = mt_dict[attribute].replace('[', '')
            mt_dict[attribute] = mt_dict[attribute].replace(']', '')
            mt_dict[attribute] = mt_dict[attribute].split(',')
            for i in range(len(mt_dict[attribute])):
                try:
                    mt_dict[attribute][i] = eval(mt_dict[attribute][i])
                except:
                    break
        else:
            mt_dict[attribute] = []
        count +=1
    missing_dict = mt_dict
    
    return missing_dict

def check_int(value):
  try:
    int(value)
    return True
  except ValueError:
    return False
```

```python
def clean_data(df):
    """
    Perform feature trimming, re-encoding, and engineering for demographics
    data
    
    INPUT: Demographics DataFrame
    OUTPUT: Trimmed and cleaned demographics DataFrame
    """
    
    # Put in code here to execute all main cleaning steps:
    count = 0
    mt_dict = {}
    for attribute in feat_info.loc[:, 'attribute']:
        mt_dict[attribute] = feat_info[feat_info['attribute'] == attribute]['missing_or_unknown'][count]
        if len(mt_dict[attribute]) > 2:
            mt_dict[attribute] = mt_dict[attribute].replace('[', '')
            mt_dict[attribute] = mt_dict[attribute].replace(']', '')
            mt_dict[attribute] = mt_dict[attribute].split(',')
            for i in range(len(mt_dict[attribute])):
                try:
                    mt_dict[attribute][i] = eval(mt_dict[attribute][i])
                except:
                    break
        else:
            mt_dict[attribute] = []
        count +=1
    na_replace = mt_dict
    # convert missing value codes into NaNs, ...
    for column in list(df.columns):
        df.loc[df[column].isin(na_replace[column]), [column]] = np.nan
    
    # remove selected columns, ...
    ## Determine which columns have enoungh NaN to be considered outliers
    nan_count = np.asarray(df.isna().sum().tolist())

    nan_mid = np.median(nan_count)
    low_mid = np.median(nan_count[nan_count < nan_mid])
    high_mid = np.median(nan_count[nan_count > nan_mid])
    iqr = high_mid - low_mid

    outliers = [low_mid - 1.5*iqr, high_mid + 1.5*iqr]
    keep_columns = []
    remove_columns =[]
    for column in df.columns:
        if outliers[0] < df[column].isna().sum() < outliers[1]:
            keep_columns.append(column)
        else:
            remove_columns.append(column)
    # outliers = ['AGER_TYP', 'GEBURTSJAHR', 'TITEL_KZ', 'ALTER_HH', 'KK_KUNDENTYP', 'KBA05_BAUMAX']
    ## Create a dataset without outlier columns
    df = df.drop(remove_columns, axis=1)
    
    # remove selected rows, ...
    # May want to revisit and apply the outlier math as before
    # 14 looked like a good place to cut the data
    df['missing_count'] = df.isna().sum(axis=1)
    df['high_nan'] = 0
    df.loc[df['missing_count'] > 14, 'high_nan'] = 1
    df = df[df['high_nan'] == 0].drop(['missing_count', 'high_nan'], axis=1)
    
    # select, re-encode, and engineer column values
    cat_var = list(feat_info[feat_info['type'] == 'categorical']['attribute'])
    mixed_var = list(feat_info[feat_info['type'] == 'mixed']['attribute'])
    # Assess categorical variables: which are binary, which are multi-level, and
    # which one needs to be re-encoded
    binary = []
    multi = []
    for column in cat_var:
        if column in list(df.columns):
            if low[column].nunique() == 2:
                binary.append(column)
            else:
                 multi.append(column)
        else:
            break
    ## Re-encode categorical variable(s) to be kept in the analysis.
    df['OST_WEST_KZ'] = df['OST_WEST_KZ'].map({'W':1, 'O':0})
    ## Impute NaN values
    def impute(df, column):
        most_freq = df[column].value_counts().idxmax()
        return df[column].fillna(most_freq)

    for var in multi:
        if var in df.keys():
            impute = impute(df, var)
            encode = pd.get_dummies(impute, prefix=var)
            df = df.drop(var, axis=1)
            for col in encode.columns:
                df[col] = encode[col]

    era_map = {key: None for key in list(range(1,16))}
    values = list(range(40,100,10))
    for i in range(1,3):
        era_map[i] = values[0]
    for i in range(3,5):
        era_map[i] = values[1]
    for i in range(5,8):
        era_map[i] = values[2]
    for i in range(8,10):
        era_map[i] = values[3]
    for i in range(10,14):
        era_map[i] = values[4]
    for i in range(14,16):
        era_map[i] = values[5]
    ag_map = {key: 0 for key in list(range(1,16))}
    ag = [2,4,6,7,9,11,13,15]
    for i in ag:
        ag_map[i] = 1
    # Create columns in dataframe relevany to decade and movement
    df['DECADE'] = df['PRAEGENDE_JUGENDJAHRE'].map(era_map)
    df['AVANTGARDE'] = df['PRAEGENDE_JUGENDJAHRE'].map(ag_map)
    # Clean the missing values if they exist:
    df['DECADE'] = impute(df, 'PRAEGENDE_JUGENDJAHRE')
    df['AVANTGARDE'] = impute(df, 'PRAEGENDE_JUGENDJAHRE')

    def wealth(digit):
        if len(str(digit)) == 2:
            tens = int(str(digit)[0])
            return tens
        else:
            return np.nan
    def life(digit):
        if len(str(digit)) == 2:
            ones = int(str(digit)[1])
            return ones
        else:
            return np.nan
    df['WEALTH'] = df['CAMEO_INTL_2015'].apply(wealth)
    df['WEALTH'] = impute(df, 'WEALTH')
    df['LIFE'] = df['CAMEO_INTL_2015'].apply(life)
    df['LIFE'] = impute(df, 'LIFE')

    rural = {key: 0 for key in list(range(1,9))}
    rural[7] = 1
    rural[8] = 1

    df['RURAL'] = df['WOHNLAGE'].map(rural)
    df['RURAL'] = impute(df, 'RURAL')
    family = {key: 1 for key in list(range(1,6))}
    family[4] = 0
    rural[5] = 0

    df['FAMILY'] = df['PLZ8_BAUMAX'].map(family)
    df['RURAL'] = impute(df, 'FAMILY')
    
    df = df.drop('LP_LEBENSPHASE_FEIN',axis=1)
    df = df.drop('LP_LEBENSPHASE_GROB',axis=1)
    
    ## Final Check
    outliers = ['AGER_TYP', 'GEBURTSJAHR', 'TITEL_KZ', 'ALTER_HH', 'KK_KUNDENTYP', 'KBA05_BAUMAX']
    dummies = ['CJT_GESAMTTYP', 'FINANZTYP','GFK_URLAUBERTYP', 'LP_FAMILIE_FEIN',
               'LP_FAMILIE_GROB', 'LP_STATUS_FEIN', 'LP_STATUS_GROB', 'NATIONALITAET_KZ',
               'SHOPPER_TYP', 'ZABEOTYP', 'GEBAEUDETYP', 'CAMEO_DEUG_2015', 
               'CAMEO_DEU_2015', 'OST_WEST_KZ']
    categoricals = ['KK_KUNDENTYP', 'TITEL_KZ', 'AGER_TYP', 'LP_LEBENSPHASE_FEIN',
                    'LP_LEBENSPHASE_GROB','WOHNLAGE', 'KBA05_BAUMAX','PLZ8_BAUMAX',
                    'PRAEGENDE_JUGENDJAHRE', 'CAMEO_INTL_2015']
    remove = outliers + dummies + categoricals
    
    for column in remove:
        if column in df.keys():
            df = df.drop(column, axis=1)
        else:
            break

    # Return the cleaned dataframe.
    return df
    
```
